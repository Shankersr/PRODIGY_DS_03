```python
# Removing outliers

column = tm[['age','campaign','duration']]
q1 = np.percentile(column, 25)
q3 = np.percentile(column, 75)
iqr = q3 - q1
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr
tm[['age','campaign','duration']] = column[(column > lower_bound) & (column < upper_bound)]
```

```python
# Plotting boxplot after removing outliers
tm.plot(kind='box', subplots=True, layout=(2,5),figsize=(20,10),color='#808000')
plt.show()
```

```python
high_corr_cols = ['emp.var.rate','euribor3m','nr.employed']
```

```python
# copy with original dataframe
tm1 = tm.copy()
tm1.columns
```

```python
# Removing high correlated columns from the dataset
tm1.drop(high_corr_cols,inplace=True,axis=1)  # axis=1 indicates columns
tm1.columns
```

```python
# showing dimensions of the updated dataset
tm1.shape
```

```python
# Conversion of categorical columns into numerical columns using label encoder.
from sklearn.preprocessing import LabelEncoder
lb = LabelEncoder()
tm_encoded = tm1.apply(lb.fit_transform)
tm_encoded
```

```python
tm_encoded['deposit'].value_counts()
```

```python
x = tm_encoded.drop('deposit',axis=1)  # independent variable
y = tm_encoded['deposit']              # dependent variable
print(x.shape)
print(y.shape)
print(type(x))
print(type(y))
```

```python
from sklearn.model_selection import train_test_split
```

```python
print(4119*0.25)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state=1)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
```

```python
def eval_model(y_test,y_pred):
    acc = accuracy_score(y_test,y_pred)
    print('Accuracy_Score',acc)
    cm = confusion_matrix(y_test,y_pred)
    print('Confusion Matrix\n',cm)
    print('Classification Report\n',classification_report(y_test,y_pred))

def mscore(model):
    train_score = model.score(x_train,y_train)
    test_score = model.score(x_test,y_test)
    print('Training Score',train_score)  # Training Accuracy
    print('Testing Score',test_score)    # Testing Accuracy
```

```python
# Importing  Decision Tree library
from sklearn.tree import DecisionTreeClassifier
```

```python
# Building Decision Tree Classifier Model
dt = DecisionTreeClassifier(criterion='gini',max_depth=5,min_samples_split=10)
dt.fit(x_train,y_train)
```

```python
# Evaluating training and testing accuracy
mscore(dt)
```

```python
# Generating prediction
ypred_dt = dt.predict(x_test)
print(ypred_dt)
```

```python
# # Evaluate the model - confusion matrix, classification Report, Accuaracy
eval_model(y_test,ypred_dt)
```
